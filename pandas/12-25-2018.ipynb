# Mapping Home_ownership

    ## Uniqueness in home_ownership

```python
ho = df['home_ownership'].unique()
ho
```

    ## mapping

```python
mapHome = {'RENT':1, 'OWN':2, 'MORTGAGE':3, 'OTHER': 4, 'NONE':5, 'ANY':6}

df['home_ownership_mapped'] = df.home_ownership.map(mapHome)
```

```python
df[['home_ownership','home_ownership_mapped']]
```

# Mapping emp_length

```python
df.emp_length.unique()
```

```python
mapEmp = {'10+ years':1,'< 1 year':2, '1 year':3, '3 years':4, '8 years':5,
          '9 years':6,'4 years':7,'5 years':8,'6 years':9, '2 years': 10, '7 years':11, 'nan': 12}

df['emp_length_mapped'] = df.emp_length.map(mapEmp)
```

```python
df[['emp_length', 'emp_length_mapped']]
```

```python
df.emp_length_mapped.isnull().sum()
```

```python
df['emp_length_mapped'].fillna(value = df['emp_length_mapped'].median(), inplace = True)
```

```python
df.emp_length_mapped.isnull().sum()
```

```python
df['emp_length_mapped']=df.emp_length_mapped.astype(int)
```
