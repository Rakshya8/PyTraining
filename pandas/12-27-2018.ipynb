```python
from sklearn import metrics
import numpy as np
from sklearn.model_selection import train_test_split
def ttr(feature):
    XX = df[feature]
    y = df.loan_condition_int
    X_train,X_test,Y_train,Y_test = train_test_split(XX,y,random_state=123)
    linreg = lr()
    linreg.fit(X_train,Y_train)
    y_pred = linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(Y_test, y_pred))
```

```python
ttr(features)
```

```python
c = [
 'loan_amount',
 'interest_rate',
 'dti',
 'grade_mapped',
]

ttr(c)
```

```python
b = [
 'loan_amount',
 'interest_rate',
 'grade_mapped',
 'total_pymnt']

ttr(b)
```
