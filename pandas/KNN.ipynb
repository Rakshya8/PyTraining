```python
import pandas as pd
```

```python
df = pd.read_csv('car.txt')
df
```

```python
df['safety'].unique()
```

```python
map_safety = {'low':10, 'med':11, 'high':12}
df['safety_mapped'] = df['safety'].map(map_safety)
df
```

```python
df['doors'].unique()
```

```python
map_doors = {'2':2, '3':3, '4':4, '5more':5}
              
df['doors_mapped'] = df['doors'].map(map_doors)
```

```python
df['persons'].unique()
```

```python
map_persons = {'2':2, '4':4, 'more':6}

df['persons_mapped'] = df['persons'].map(map_persons)
```

```python
x = df[['doors_mapped','persons_mapped']]
y = df['safety_mapped']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train) 
```

```python
neigh.score(X_test,y_test)
```

```python
pred = neigh.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, pred.round())
accuracy
```
