# Comparing population between Zimbawe and Algeria

```python
zm = df[df.country == 'Zimbabwe']
zm
```

```python
alg = df[df.country == 'Algeria']
alg
```

```python
pt.plot(zm.year, zm.population/10**6)
pt.plot(alg.year, alg.population/10**6)
pt.xlabel('Year')
pt.ylabel('Population in millions')
pt.title('Population growth in Zimbawe and Algeria')
pt.legend(['Zimbawe','Algeria'])
pt.show()
```
